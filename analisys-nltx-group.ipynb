```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
```

```python
all_indicators_df = pd.read_csv('body/wmd_all_indicators.csv')
```

```python
not_arg_df = all_indicators_df[all_indicators_df['label'] == 'nao_argumento']
arg_df = all_indicators_df[all_indicators_df['label'] != 'nao_argumento']

not_arg_rates = not_arg_df['wmd_rate']
arg_rates = arg_df['wmd_rate']
```

```python
f = px.box(all_indicators_df, y='wmd_rate', x='label', title="Nota do WMD por tipos de sentenÃ§as")
f.update_yaxes(
    dtick=0.009
)
f.show()
```

```python
grouped_by_arg = all_indicators_df.copy()
grouped_by_arg['label'] = all_indicators_df['label'].apply(lambda x: x if x == 'nao_argumento' else 'argumento')
```

```python
f = px.box(grouped_by_arg, y='wmd_rate', color='label', notched=True, color_discrete_sequence=['forestgreen', 'indianred'], title="Nota do WMD - Argumento x NÃ£o argumento", height=610)
f.update_yaxes(
    dtick=0.006
)
f.show()
```

# Histograma
Tamanho das redaÃ§Ãµes: quantidade de palavras

```python
all_indicators_df['len_text'] = all_indicators_df['text'].apply(lambda x: len(x.split()))
grouped_by_arg['len_text'] = grouped_by_arg['text'].apply(lambda x: len(x.split()))
```

```python
np.arange(0, 71, 10)
```

```python
fig = px.histogram(all_indicators_df, x='len_text', color="label", nbins=10, labels={"len_text": "quantidade de palavras"}, title="Histograma de quantidade de palavras dos trechos de redaÃ§Ã£o", barmode='group')
fig.update_xaxes(
    showgrid=True,
    dtick = 10
)
fig.show()
```

```python
fig = px.histogram(grouped_by_arg, x='len_text', color="label", labels={"len_text": "quantidade de palavras"}, color_discrete_sequence=["limegreen", "indianred"], nbins=10, barmode='group', title="Comparativo entre redaÃ§Ãµes marcadas como argumento e nÃ£o argumento")

fig.update_xaxes(
    showgrid=True,
    dtick = 10
)
fig.show()
```

## histograma da base do nlx-group

```python

nlx_group_df = pd.read_csv('data-arguments/data-nlx-group.csv')
nlx_group_df['len_text'] = nlx_group_df['text'].apply(lambda x: len(x.split()))
```

```python
print("Quantidade de sentenÃ§as cujo nÃºmero de palavras Ã© maior ou igual a 80:\n", nlx_group_df[nlx_group_df['len_text'] >= 80].shape[0])

print("Quantidade de sentenÃ§as cujo nÃºmero de palavras Ã© menor que 80:\n", nlx_group_df[nlx_group_df['len_text'] < 80].shape[0])
```

Irei remover todas as sentenÃ§as acima de 80 palavras, pois atrapalha na visualizaÃ§Ã£o do histograma. Tem algumas sentenÃ§as de 600 palavras, outras de 200. Somando todas as sentenÃ§as acima de 80 palavras dÃ¡ apenas 99 sentenÃ§as.

```python
nlx_group_df = nlx_group_df[nlx_group_df['len_text'] < 80]
```

```python
fig = px.histogram(nlx_group_df, x='len_text', color="label", labels={"len_text": "quantidade de palavras"}, color_discrete_sequence=["red", "limegreen"], nbins=10, barmode='group', title="Comparativo entre textos do nlx-group marcados como argumento e nÃ£o argumento")

fig.update_xaxes(
    showgrid=True,
    dtick = 10
)
fig.show()
```
